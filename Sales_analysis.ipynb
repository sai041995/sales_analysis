```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python
### clean null values
```

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()
all_data = all_data.dropna(how="all")
all_data.head()
```

```python
### lets remove all above null values
```

```python
all_data = all_data.dropna(how="all")
all_data.head()
```

```python
###find some errors and delete it
```

```python
temp_df = all_data[all_data["Order Date"].str[0:2] == "Or"]
temp_df.head()
```

```python
### lets delete all or 
```

```python
all_data = all_data[all_data["Order Date"].str[0:2] != "Or"]
all_data.head()
```

```python
### Convert coloumns in to correct 
```

```python
all_data["Quantity Ordered"] = pd.to_numeric(all_data["Quantity Ordered"]) # INT
all_data["Price Each"] = pd.to_numeric(all_data["Price Each"]) # FLOAT
```

```python
### Task-0 sorting dataframe
### Add monthh colounm
```

```python
all_data["Month"] = all_data["Order Date"].str[0:2]
all_data["Month"] = all_data["Month"].astype("int32")


all_data.head()
```

```python
### Task-3 Add sales coloumn
```

```python
all_data["Sales"] = all_data["Quantity Ordered" ] * all_data["Price Each"]
all_data.head()
```

```python
### analysis data begins from here
### Task-2 what was the best month for sales , and how much was earned?
```

```python
all_data.groupby("Month").sum()
```

```python
### lets plot this
```

```python
Results = all_data.groupby("Month").sum()
months = range(1,13)
plt.bar(months, Results["Sales"])
plt.xticks(months)
plt.ylabel("Sales in USD ($)")
plt.xlabel("Months")
plt.show
```

```python
### Task-3 add a city coloumn by using .apply methode
```

```python
def get_city(address):
    return address.split(",")[1]

def get_state(address):
    return address.split(",")[2].split(" ")[1]

all_data["City"] = all_data["Purchase Address"].apply(lambda x : f"{get_city(x)}  ({get_state(x)})")

all_data.head()
```

```python
### Task-4  what us city have highst number of sales
```

```python
all_data.groupby("City").sum()
```

```python
### lets plot this
```

```python
results = all_data.groupby("City").sum()
results
```

```python
cities = [city for city , df in all_data.groupby("City")]

plt.bar(cities,results["Sales"])
plt.xticks(cities,rotation="vertical",size=8)
plt.ylabel("Sales in USD ($)")
plt.xlabel("Cities")
plt.show()
```

```python
### lets split and correct the orderdate

all_data.head()
```

```python
all_data["Order Date"] = pd.to_datetime(all_data["Order Date"])
all_data.head()
```

```python
all_data["Hour"] = all_data["Order Date"].dt.hour
all_data["Minute"] = all_data["Order Date"].dt.minute
all_data["Count"] = 1
all_data.head()
```

```python
### Task-4 what time is best to advertise to increase in customers buying their likely product?
```

```python
Hours = [Hour for Hour, df in all_data.groupby("Hour")] 

all_data.groupby(["Hour"]).count()
```

```python
### lets plot this
```

```python
Hours = [Hour for Hour, df in all_data.groupby("Hour")]
plt.plot(Hours,all_data.groupby(["Hour"]).count())
plt.xticks(Hours)
plt.grid()
plt.xlabel("Hours")
plt.ylabel("Orders")
plt.show()
```
